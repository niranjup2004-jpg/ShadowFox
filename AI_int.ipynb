```python
!pip install -q seaborn pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
discount_profit_corr = df[['Discount', 'Profit']].corr().iloc[0, 1]
print(f"âœ… Total Sales: ${total_sales:,.2f}")
print(f"âœ… Total Profit: ${total_profit:,.2f}")
print(f"ðŸ“‰ Correlation between Discount and Profit: {discount_profit_corr:.2f}")
category_perf = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
subcategory_perf = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Profit', ascending=False)
region_perf = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Profit', ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=category_perf.index, y=category_perf['Sales'], palette='Blues_d')
plt.title('ðŸ’° Total Sales by Category')
plt.ylabel('Sales ($)')
plt.xlabel('Product Category')
plt.tight_layout()
plt.show()
plt.figure(figsize=(12, 5))
sns.barplot(x=subcategory_perf.index, y=subcategory_perf['Profit'], palette='Greens_d')
plt.title('ðŸ“Š Profit by Sub-Category')
plt.ylabel('Profit ($)')
plt.xlabel('Sub-Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 5))
sns.barplot(x=region_perf.index, y=region_perf['Profit'], palette='Oranges_d')
plt.title('ðŸ“ Profit by Region')
plt.ylabel('Profit ($)')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```
